```python
import pandas as pd
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Luas Panen, Produksi, dan Produktivitas Padi Menurut Provinsi, 2023.xlsx', skiprows=3)
data.head()
```

```python
# Rename columns
data.columns = ['Provinsi', 'Luas Panen (ha)', 'Produktivitas (ku/ha)', 'Produksi (ton)']

# Hapus baris yang mengandung NaN
data = data.dropna()

# Pastikan kolom numerik bertipe numerik
data['Luas Panen (ha)'] = pd.to_numeric(data['Luas Panen (ha)'], errors='coerce')
data['Produktivitas (ku/ha)'] = pd.to_numeric(data['Produktivitas (ku/ha)'], errors='coerce')
data['Produksi (ton)'] = pd.to_numeric(data['Produksi (ton)'], errors='coerce')

# Hapus baris yang masih memiliki NaN setelah konversi
data = data.dropna()

data.head()
```
